```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})
data.head()

```

```python
lst_robot = []
lst_human = []
for i in range(len(data['whoAmI'])):
  if data['whoAmI'][i] == 'robot':
    lst_robot.append(1)
    lst_human.append(0)
  else:
    lst_robot.append(0)
    lst_human.append(1)

new_data = pd.DataFrame({'robot': lst_robot, 'human': lst_human})
new_data.head()
```
